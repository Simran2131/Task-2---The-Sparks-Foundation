```python
data['species'].value_counts()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc1=dc.fit(x_train,y_train)
```

```python
from sklearn import tree
```

```python
## WE Can Visualise Decision Tree of Iris Dataset as Follows :-
```

```python
plt.figure(figsize=(25,17))
t=tree.plot_tree(dc1,feature_names=['sepal_length','sepal_width','petal_length','petal_width'],
                class_names=['Setosa','Verginica','Versicolor'],
                filled=True)
```

# Prediction on data

```python
y_pred=dc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred)*100)
```
